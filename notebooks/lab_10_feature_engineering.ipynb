## Parte 1: Exemplo Guiado - O Poder do `.assign()`
Para não poluir nosso código com variáveis temporárias (`df['nova'] = ...`), usamos o método `.assign()` do Pandas com funções `lambda`. Ele permite encadear a criação de múltiplas colunas de forma elegante e segura.

```
df_bruto = pd.DataFrame({
    'Produto': ['A', 'B', 'C'],
    'Preco_Venda': [120.0, 250.0, 180.0],
    'Custo_Fabrica': [50.0, 100.0, 80.0]
})

# Feature Engineering Funcional (Method Chaining)
df_enriquecido = (
    df_bruto
    .assign(
        Lucro = lambda x: x['Preco_Venda'] - x['Custo_Fabrica'],
        Margem_Percentual = lambda x: (x['Lucro'] / x['Preco_Venda']) * 100,
        Produto_Premium = lambda x: (x['Preco_Venda'] > 150).astype(int)
    )
)
display(df_enriquecido)
```

## Parte 2: Os Engenheiros Financeiros
Você recebeu um *dump* de um e-commerce contendo as métricas de clientes.
**Sua Tarefa:** Utilizando estritamente o método `.assign()` com funções `lambda` em uma única cadeia de execução:
1. Calcule o **Lucro_Total** (`Receita_Total` - `Custo_Total`).
2. Calcule o **Ticket_Medio** (`Receita_Total` dividida por `Qtd_Compras`).
3. Crie a Flag **Is_VIP** (clientes que tiveram `Receita_Total` estritamente maior que R$ 1.500).

```
np.random.seed(10)
df_ecommerce = pd.DataFrame({
    'ID_Cliente': range(100, 105),
    'Receita_Total': np.random.uniform(500, 3000, 5),
    'Custo_Total': np.random.uniform(200, 800, 5),
    'Qtd_Compras': np.random.randint(1, 10, 5)
})
```

```
# ESCREVA SEU CÓDIGO AQUI
# Resolva utilizando encadeamento: df_final = df_ecommerce.assign(...)

df_final = (
    df_ecommerce.assign(
        # 1. Calcule o Lucro_Total
        Lucro_Total=lambda x: x['Receita_Total'] - x['Custo_Total'],

        # 2. Calcule o Ticket_Medio
        Ticket_Medio=lambda x: x['Receita_Total'] / x['Qtd_Compras'],

        # 3. Crie a flag Is_VIP (use .astype(int))
        Is_VIP=lambda x: (x['Receita_Total'] > 1500).astype(int)
    )
)

display(df_final)
```
